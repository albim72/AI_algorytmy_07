```python
import tensorflow as tf
```

```python
mnist = tf.keras.datasets.mnist
```

```python
(x_train,y_train),(x_test,y_test) = mnist.load_data()
```

```python
x_train,x_test = x_train/255.0,x_test/255.0
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#przygtowanie modelu prostej sieci neuronowej
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(128,activation='relu'), #parametry = 784x128+128 = 100 480
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(10)
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics = ["accuracy"])
```

```python
#ocena niwytrenowanego modelu
test_loss,test_acc = model.evaluate(x_test,y_test)
print(f"strata niewytrenowany: {test_loss}")
print(f"acc niewytrenowany: {test_acc}")
```

```python
model.fit(x_train,y_train, epochs=10)
```

```python
#ocena modelu
test_loss,test_acc = model.evaluate(x_test,y_test)
print(f"strata: {test_loss}")
print(f"acc: {test_acc}")
```
